```python
#program to extract and display all the header tags from en.wikipedia.org/wiki/Main_Page
from urllib.request import urlopen
from bs4 import BeautifulSoup
html = urlopen('https://en.wikipedia.org/wiki/Main_Page')
bs = BeautifulSoup(html, "html.parser")
titles = bs.find_all(['h1', 'h2','h3','h4','h5','h6'])
print('List all the header tags :', *titles, sep='\n\n')
```

```python
#program to check a string contains set of characters
import re
def is_allowed_specific_char(str):
    #charRe = re.compile(r'[a-z]')
    #string = charRe.search(str)
    #return bool (string)
    result = re.match(r'[a-zA-Z0-9]', str)
    return bool(result)
    
  
print(is_allowed_specific_char("abcdefABCDEF012345")) 
print(is_allowed_specific_char("*&%@#!}{"))
```

```python
#program to match a string has a followed by zero or more b's
import re
def text_match(text):
        patterns = '^(a(b*)$)'
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')
print(text_match("bc"))
print(text_match("abc"))
print(text_match("a"))
print(text_match("ab"))
print(text_match("abb"))
```

```python
#program to match a string has a followed by one or more b's
import re
def text_match(text):
        patterns = 'ab+$'
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("ab"))
print(text_match("abc"))
print(text_match("abbbb"))
print(text_match("aaaab"))
print(text_match("abbabbaab"))
print(text_match("abbbababb"))
```

```python
#program to match a string has a followed by zero or one 'b'
import re
def text_match(text):
        patterns = 'ab?$'
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')
print(text_match("a"))
print(text_match("ab"))
print(text_match("abc"))
print(text_match("abbc"))
print(text_match("aabbc"))
```

```python
#program that matches a string that has an a followed by three 'b'
import re
def text_match(text):
        patterns = 'ab{3}?$'
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("abbb"))
print(text_match("aabbbbbc"))
```

```python
#program to find sequences of lowercase letters joined with a underscore
import re
def text_match(text):
        patterns = '^[a-z]+_[a-z]+$'
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("aab_cbbbc"))
print(text_match("aab_Abbbc"))
print(text_match("Aaab_abbbc"))
```
